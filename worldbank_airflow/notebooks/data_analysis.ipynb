```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
BASE_PATH = '/opt/airflow/data'
RAW_PATH = os.path.join(BASE_PATH, 'raw')
PROCESSED_PATH = os.path.join(BASE_PATH, 'processed')
MERGED_PATH = os.path.join(BASE_PATH, 'merged')
```

```python
graphs_folder = Path('worldbank_airflow/data/graphs')
graphs_folder.mkdir(parents=True, exist_ok=True)
```

```python
unemployment_CZ = pd.read_csv(
    '/Users/arortega/Desktop/git/world_bank_etl/worldbank_airflow/data/merged/CZ_merged.csv'
)

unemployment_EU = pd.read_csv(
    '/Users/arortega/Desktop/git/world_bank_etl/worldbank_airflow/data/merged/EUU_merged.csv'
)

unemployment_merged_df = pd.concat([unemployment_CZ, unemployment_EU], axis=0)
```

```python
indicators = [
    indicator for indicator in unemployment_merged_df.columns if indicator not in ['country', 'year']
]
```

```python
indicators_dict = {
    'total_unemployment_rate': 'Total Unemployment Rate',
    'unemployment_rate_advanced_education_male': 'Unemployment Rate Advanced Education Male',
    'unemployment_rate_advanced_education_female': 'Unemployment Rate Advanced Education Female',
    'youth_unemployment_rate': 'Youth Unemployment Rate',
    'unemployment_rate_intermediate_education_male': 'Unemployment Rate Intermediate Education Male',
    'male_unemployment_rate': 'Male Unemployment Rate',
    'unemployment_rate_basic_education_male': 'Unemployment Rate Basic Education Male',
    'unemployment_rate_basic_education_female': 'Unemployment Rate Basic Education Female',
    'unemployment_rate_intermediate_education_female': 'Unemployment Rate Intermediate Education Female',
    'female_unemployment_rate': 'Female Unemployment Rate',
    'youth_male_unemployment_rate': 'Youth Male Unemployment Rate',
    'youth_female_unemployment_rate': 'Youth Female Unemployment Rate'
}
```

```python
for indicator_technical, indicator_text in indicators_dict.items():
    plt.figure(figsize=(10, 6))

    sns.lineplot(
        data=unemployment_merged_df
        , x='year'
        , y=indicator_technical
        , hue='country'
        , marker='o'
    )

    plt.title(f'{indicator_text} (%) over Years by Country')
    plt.xlabel('Year')
    plt.ylabel(f'{indicator_text} (%)')
    plt.legend(title='Country')
    plt.grid(True)

    # plt.show()

    plt.savefig(
        f'{graphs_folder}/{indicator_technical}_CZ_EUU.png'
        , dpi=300
        , bbox_inches='tight'
    )

```
